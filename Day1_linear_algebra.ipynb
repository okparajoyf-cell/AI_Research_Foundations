```python
import numpy as np

# Scalar
s = np.array(5)
print(f"Scalar: {s}, Shape: {s.shape}")

# Vector (1D Tensor)
v = np.array([1, 2, 3])
print(f"Vector: {v}, Shape: {v.shape}")

# Matrix (2D Tensor)
m = np.array([[1, 2], [3, 4]])
print(f"Matrix:\n{m}, Shape: {m.shape}")
```
